```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Data collection and processing

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Spilliting the data into trainning data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model trainning Logistic regreesion

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
# Generate time vector
fs = 1000  # Sampling frequency
T = 10     # Duration in seconds
t = np.arange(0, T, 1/fs)

# Generate ECG signal (example)
f1 = 0.5  # Frequency of first component (P wave)
f2 = 1.0  # Frequency of second component (QRS complex)
f3 = 0.25 # Frequency of third component (T wave)
f4 = 0.1   # Frequency of fourth component (P-R segment)
f5 = 0.2   # Frequency of fifth component (ST segment)
ecg_signal = (
    np.sin(2*np.pi*f1*t) +  # P wave
    np.sin(2*np.pi*f2*t) +  # QRS complex
    0.5*np.sin(2*np.pi*f3*t)  # T wave
)

# Add P-R segment
pr_segment = 0.3 * np.sin(2*np.pi*f4*t)
ecg_signal += pr_segment

# Add ST segment
st_segment = 0.2 * np.sin(2*np.pi*f5*t)
ecg_signal += st_segment

# Add noise to the signal
noise = 0.1 * np.random.normal(size=len(t))
ecg_with_noise = ecg_signal + noise

# Plot the ECG waveform
plt.figure(figsize=(8, 4))
plt.plot(t, ecg_with_noise, color='black')
plt.title('ECG Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
